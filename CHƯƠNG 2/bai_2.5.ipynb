```python
import xml.etree.ElementTree as ET
try:
    tree = ET.parse("sample.xml")
    root = tree.getroot()
    id_delete = input("Nhập ID nhân viên cần xoá: ").strip()
    found = False
    for staff in root.findall("staff"):
        if staff.get("id") == id_delete:
            root.remove(staff)
            found = True
            break
    if found:
        print(f" Đã xoá nhân viên có ID = {id_delete}")
    else:
        print(f" Không tìm thấy nhân viên có ID = {id_delete}")
    tree.write("delete_sample.xml", encoding="utf-8", xml_declaration=True)
    print("\n===== DANH SÁCH NHÂN VIÊN SAU KHI XOÁ =====")
    for staff in root.findall("staff"):
        id_nv = staff.get("id")
        name = staff.find("name").text
        salary = staff.find("salary").text
        print(f"ID: {id_nv}")
        print(f"Tên: {name}")
        print(f"Lương: {salary}")
        print("-----------------------------")
except FileNotFoundError:
    print(" Lỗi: Không tìm thấy file 'sample.xml'. Hãy kiểm tra lại đường dẫn.")
except ET.ParseError:
    print(" Lỗi: File XML bị sai cấu trúc. Hãy kiểm tra lại nội dung.")
```
