```python
data = {
    "Công ty": "Công ty TNHH Đất Việt",
    "Địa chỉ": "abc Giải Phóng – Hà Nội",
    "Nhân viên": [
        {"Tên": "A", "Đơn vị": "A1"},
        {"Tên": "B", "Đơn vị": "A1"},
        {"Tên": "C", "Đơn vị": "A2"},
        {"Tên": "D", "Đơn vị": "A3"},
        {"Tên": "E", "Đơn vị": "A3"}
    ]
}
tong_nv = len(data["Nhân viên"])
thongke = {}
for nv in data["Nhân viên"]:
    dv = nv["Đơn vị"]
    thongke[dv] = thongke.get(dv, 0) + 1
print(f"Tên công ty: {data['Công ty']}")
print(f"Địa chỉ: {data['Địa chỉ']}")
print("----- Thống kê nhân viên -----")
for dv, sl in thongke.items():
    tyle = sl / tong_nv * 100
    print(f"Đơn vị {dv}: {sl} nhân viên - Tỷ lệ: {tyle:.2f}%")
```
