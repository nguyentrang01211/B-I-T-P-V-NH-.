```python
import xml.etree.ElementTree as ET
root = ET.Element("dssinhvien")
while True:
    sv = ET.Element("sinhvien")
    masv = input("Nhập mã sinh viên: ")
    hoten = input("Nhập họ tên: ")
    namsinh = input("Nhập năm sinh: ")
    lop = input("Nhập lớp: ")
    gt = input("Giới tính: ")
    ET.SubElement(sv, "masv").text = masv
    ET.SubElement(sv, "hoten").text = hoten
    ET.SubElement(sv, "namsinh").text = namsinh
    ET.SubElement(sv, "lop").text = lop
    ET.SubElement(sv, "gioitinh").text = gt
    root.append(sv)
    cont = input("Nhập thêm sinh viên khác? (c/k): ")
    if cont.lower() == "k":
        break
tree = ET.ElementTree(root)
tree.write("sinhvien.xml", encoding="utf-8", xml_declaration=True)
print("✅ Đã lưu file sinhvien.xml thành công!")
```
