```python
import xml.etree.ElementTree as ET

try:
    tree = ET.parse("sample.xml")
    root = tree.getroot()
    new_staff = ET.Element("staff", id="3") 
    name = ET.SubElement(new_staff, "name")
    name.text = "Anthony Walter"
    salary = ET.SubElement(new_staff, "salary")
    salary.text = "3.2 LPA"
    root.append(new_staff)
    tree.write("new_sample.xml", encoding="utf-8", xml_declaration=True)
    print("===== DANH SÁCH NHÂN VIÊN SAU KHI THÊM =====")
    for staff in root.findall("staff"):
        id_nv = staff.get("id")
        name = staff.find("name").text
        salary = staff.find("salary").text
        print(f"ID: {id_nv}")
        print(f"Tên: {name}")
        print(f"Lương: {salary}")
        print("-----------------------------")
except FileNotFoundError:
    print(" Lỗi: Không tìm thấy file 'sample.xml'. Hãy kiểm tra lại đường dẫn.")
except ET.ParseError:
    print(" Lỗi: File XML bị sai cấu trúc. Hãy kiểm tra lại nội dung.")

```
