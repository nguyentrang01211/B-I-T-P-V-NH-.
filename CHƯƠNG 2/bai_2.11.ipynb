```python
import json
import datetime

def nhap_gd():
    ds = []
    while True:
        ten = input("Tên giao dịch: ")
        so = float(input("Số tiền: "))
        loai = input("Loại (thu/chi): ")
        ds.append({"Tên": ten, "Số tiền": so, "Loại": loai})
        if input("Nhập thêm? (c/k): ").lower() == 'k':
            break
    return ds

def luu_json(ds):
    now = datetime.datetime.now()
    filename = now.strftime("%Y-%m-%d-%H-%M-%S.json")
    with open(filename, "w", encoding="utf-8") as f:
        json.dump(ds, f, indent=4, ensure_ascii=False)
    print(f"✅ Đã lưu file {filename}")

ds = nhap_gd()
chon = input("Lưu file? (1=có, 0=không): ")
if chon == "1":
    luu_json(ds)


```
