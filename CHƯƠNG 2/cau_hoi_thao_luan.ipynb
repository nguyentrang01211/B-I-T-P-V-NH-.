1.XML là ngôn ngữ đánh dấu mở rộng được thiết kế để lưu trữ và trao đổi dữ liệu một cách có cấu trúc, dễ hiểu cho cả người và máy.

2.DOM (Document Object Model) là mô hình biểu diễn tài liệu XML như một cây (tree), trong đó:
Mỗi phần tử XML là một node (nút)
Có root node (gốc), element nodes, text nodes, attribute nodes

3.Cú pháp tài liệu XMl là
<?xml version="1.0" encoding="UTF-8"?>
<root_element>
    <child_element attribute="value">Nội dung</child_element>
</root_element>


4.JSON là định dạng dữ liệu dạng văn bản (text) dùng để lưu trữ và trao đổi dữ liệu giữa các ứng dụng.

5.Hạn chế của việc khai thác JSON trong python là
JSON chỉ hỗ trợ một số kiểu dữ liệu cơ bản
Không lưu được đối tượng phức tạp (class, function, object)
Dữ liệu số lớn dễ mất độ chính xác
Dữ liệu chỉ ở dạng chuỗi
Không có chú thích (comments)
Dễ bị lỗi khi format sai
Không hỗ trợ tham chiếu hoặc đối tượng lồng phức tạp

6.API XML là một loại API trả dữ liệu về dưới dạng XML (thay vì JSON).


7.API JSON (JSON API) là một giao diện lập trình ứng dụng (API) mà trong đó dữ liệu được trao đổi giữa client và server ở định dạng JSON (JavaScript Object Notation).
