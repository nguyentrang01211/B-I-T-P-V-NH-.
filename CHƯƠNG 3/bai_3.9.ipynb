```python
import pandas as pd

euro = pd.read_csv("euro2012.csv")

print(euro["Goals"])

print("Số đội:", euro["Team"].nunique())

discipline = euro[["Team", "Yellow Cards", "Red Cards"]]
discipline_sorted = discipline.sort_values(["Red Cards", "Yellow Cards"], ascending=False)

print("Mean Yellow Cards:", euro["Yellow Cards"].mean())

print("Teams > 6 goals:")
print(euro[euro["Goals"] > 6])

print("Teams start with G:")
print(euro[euro["Team"].str.startswith("G")])

print(euro.iloc[:, :7])
print(euro.iloc[:, :-3])

cols = ["Team","Goals","Shooting Accuracy","Yellow Cards","Red Cards"]
print(euro[cols])

print(euro[euro["Team"].isin(["England","Italy","Russia"])][["Team","Shooting Accuracy"]])
```
