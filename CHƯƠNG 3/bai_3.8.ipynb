```python
import numpy as np

positions = [x.strip() for x in open("positions.txt")]
heights    = np.array([float(x.strip()) for x in open("heights.txt")])

np_positions = np.array(positions)
np_heights = heights

print(np_positions.dtype, np_heights.dtype)

# GK
print("Mean height GK:", np_heights[np_positions == "GK"].mean())

# structured array
players = np.zeros(len(np_positions), dtype=[("position","U5"),("height","f4")])
players["position"] = np_positions
players["height"] = np_heights

ordered = np.sort(players, order="height")
print("Shortest:", ordered[0])
print("Tallest:", ordered[-1])
```
