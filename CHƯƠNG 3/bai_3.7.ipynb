```python
import numpy as np

with open("baseball_2D.txt") as f:
    baseball = [list(map(float, line.split())) for line in f]

np_baseball = np.array(baseball)

print("Type:", type(np_baseball))
print("Shape:", np_baseball.shape)

print("Row 50:", np_baseball[49])

np_weight = np_baseball[:,1]

print("Height of player 124:", np_baseball[123,0])

print("Mean height:", np_baseball[:,0].mean())
print("Mean weight:", np_baseball[:,1].mean())

corr = np.corrcoef(np_baseball[:,0], np_baseball[:,1])[0,1]
print("Correlation height-weight:", corr)
```
