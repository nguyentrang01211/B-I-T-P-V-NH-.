```python
import numpy as np
# 1. tạo arr_zeros có 10 phần tử 0, cập nhật phần tử ở vị trí thứ 5 là 1
arr_zeros = np.zeros(10, dtype=int)
arr_zeros[4] = 1   # index 4 tương ứng vị trí thứ 5
print(arr_zeros)

# 2. tạo arr_h có giá trị từ 10 đến 24, in danh sách các phần tử theo thứ tự đảo ngược
arr_h = np.arange(10, 25)
print("arr_h:", arr_h)
print("arr_h reversed:", arr_h[::-1])

# 3. Cho arr_k = np.array([1,2,0,8,2,0,1,3,0,5,0]) tạo arr_l từ arr_k với các phần tử khác 0
arr_k = np.array([1,2,0,8,2,0,1,3,0,5,0])
arr_l = arr_k[arr_k != 0]
print("arr_l (không phải 0):", arr_l)

# 4. Từ arr_l của câu 3, thêm 2 phần tử có giá trị 10 và 20 vào cuối array
arr_l_app = np.append(arr_l, [10,20])
print("arr_l after append:", arr_l_app)

# 5. Từ array của câu 4, thêm phần tử có giá trị 100 vào vị trí index = 5
idx = 5
arr_with_100 = np.insert(arr_l_app, idx, 100)
print("after insert 100 at index 5:", arr_with_100)

# 6. Từ array của câu 5, xóa các phần tử ở vị trí index = 0,1,2
arr_removed = np.delete(arr_with_100, [0,1,2])
print("after delete indices 0,1,2:", arr_removed)
```
