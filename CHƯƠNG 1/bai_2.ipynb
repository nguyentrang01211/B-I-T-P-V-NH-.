```python
class ThiSinh:
    def __init__(self, ho_ten="", toan=0, ly=0, hoa=0):
        self.ho_ten = ho_ten
        self.toan = toan
        self.ly = ly
        self.hoa = hoa

    def nhap(self):
        self.ho_ten = input("Nháº­p há» tÃªn thÃ­ sinh: ")
        self.toan = float(input("Äiá»ƒm ToÃ¡n: "))
        self.ly = float(input("Äiá»ƒm LÃ½: "))
        self.hoa = float(input("Äiá»ƒm HÃ³a: "))

    def tong_diem(self):
        return self.toan + self.ly + self.hoa

    def hien_thi(self):
        print(f"{self.ho_ten} - ToÃ¡n: {self.toan}, LÃ½: {self.ly}, HÃ³a: {self.hoa}, Tá»•ng: {self.tong_diem()}")
if __name__ == "__main__":
    ds = []
    n = int(input("Nháº­p sá»‘ lÆ°á»£ng thÃ­ sinh: "))

    # Nháº­p danh sÃ¡ch thÃ­ sinh
    for i in range(n):
        print(f"\n--- Nháº­p thÃ­ sinh {i+1} ---")
        ts = ThiSinh()
        ts.nhap()
        ds.append(ts)

    # In danh sÃ¡ch
    print("\nðŸ“Œ Danh sÃ¡ch thÃ­ sinh:")
    for ts in ds:
        ts.hien_thi()

    # In thÃ­ sinh trÃºng tuyá»ƒn
    print("\nâœ… Danh sÃ¡ch thÃ­ sinh trÃºng tuyá»ƒn (Tá»•ng â‰¥ 15):")
    for ts in ds:
        if ts.tong_diem() >= 15:
            ts.hien_thi()
```
