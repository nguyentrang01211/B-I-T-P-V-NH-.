1. Đóng gói (Encapsulation)
-Gộp dữ liệu (thuộc tính) và phương thức (hành vi) vào chung một đối tượng.
-Giúp ẩn thông tin: chỉ cho phép truy cập dữ liệu thông qua các phương thức, tránh sửa trực tiếp → tăng tính an toàn.
2. Kế thừa (Inheritance)
-Cho phép lớp con (class con) kế thừa thuộc tính và phương thức của lớp cha.
-Giúp tái sử dụng code, dễ mở rộng chương trình.
3. Đa hình (Polymorphism)
-Cùng một tên phương thức nhưng hành vi khác nhau tùy đối tượng/lớp.
-Giúp linh hoạt, dễ mở rộng (ví dụ: phương thức draw() có thể vẽ hình tròn, hình vuông, hình tam giác khác nhau).
4. Trừu tượng (Abstraction)
-Ẩn đi chi tiết cài đặt, chỉ giữ lại những đặc điểm quan trọng.
-Giúp người dùng làm việc với đối tượng dễ dàng mà không cần biết nội bộ nó hoạt động thế nào.
