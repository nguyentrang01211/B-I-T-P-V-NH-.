**Các loại thừa kế trong lập trình hướng đối tượng 
1. Kế thừa đơn (Single Inheritance)
-Một lớp con chỉ kế thừa từ một lớp cha duy nhất.
-Đơn giản, dễ quản lý.

```python
class Animal:
    pass
class Dog(Animal):   # Dog kế thừa Animal
    pass
```

2. Kế thừa đa cấp (Multilevel Inheritance)
-Một lớp con kế thừa từ lớp cha, và lớp cha đó lại kế thừa từ lớp cha khác.
-Tạo thành chuỗi kế thừa nhiều tầng.

```python
class Animal:
    pass
class Mammal(Animal):   # Mammal kế thừa Animal
    pass
class Dog(Mammal):      # Dog kế thừa Mammal
    pass
```

3. Kế thừa phân cấp (Hierarchical Inheritance)
-Nhiều lớp con kế thừa cùng một lớp cha.

```python
class Animal:
    pass
class Dog(Animal):      # Dog kế thừa Animal
    pass
class Cat(Animal):      # Cat kế thừa Animal
    pass
```

4. Kế thừa đa (Multiple Inheritance)
-Một lớp con kế thừa từ nhiều lớp cha cùng lúc.
-Có thể gây mâu thuẫn (ví dụ: 2 lớp cha có cùng phương thức), phải giải quyết bằng quy tắc (Python dùng MRO – Method Resolution Order).

```python
class A:
    pass
class B:
    pass
class C(A, B):   # C kế thừa từ cả A và B
    pass
```

5.Kế thừa lai (Hybrid Inheritance)
-Kết hợp nhiều loại kế thừa trên trong cùng một chương trình.
-Phức tạp, dễ gây mâu thuẫn, cần cơ chế xử lý.
Ví dụ: một lớp vừa kế thừa đa cấp, vừa kế thừa đa.
