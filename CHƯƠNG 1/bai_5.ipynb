```python
class Stack:
    def __init__(self, size=10):
        self.size = size
        self.stack = []

    def push(self, value):
        """Thêm phần tử vào ngăn xếp"""
        if not self.isFull():
            self.stack.append(value)
            print(f"Đã push {value} vào stack.")
        else:
            print("❌ Ngăn xếp đã đầy!")

    def pop(self):
        """Lấy phần tử trên cùng ra"""
        if not self.isEmpty():
            value = self.stack.pop()
            print(f"Đã pop {value} khỏi stack.")
            return value
        else:
            print("❌ Ngăn xếp rỗng!")
            return None

    def isEmpty(self):
        return len(self.stack) == 0

    def isFull(self):
        return len(self.stack) == self.size

    def count(self):
        """Trả về số phần tử hiện có"""
        return len(self.stack)

    def __str__(self):
        return f"Stack ({self.count()}/{self.size}): {self.stack}"
if __name__ == "__main__":
    s = Stack(5)

    s.push(10)
    s.push(20)
    s.push(30)
    print(s)
    print("Số phần tử trong stack:", s.count())

    s.pop()
    print(s)
    print("Số phần tử trong stack:", s.count())

    s.push(40)
    s.push(50)
    s.push(60)
    print(s)
    print("Số phần tử trong stack:", s.count())
    
```
