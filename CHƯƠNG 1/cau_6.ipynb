-Lớp trừu tượng là một lớp không thể tạo đối tượng trực tiếp.
-Nó được dùng làm khuôn mẫu (blueprint) cho các lớp con.
-Trong lớp trừu tượng có thể chứa:
-Phương thức trừu tượng: chỉ khai báo tên, không có phần thân, bắt buộc lớp con phải định nghĩa lại.
-Phương thức bình thường: có thể cài đặt sẵn, lớp con kế thừa dùng luôn.
**Đặc điểm 
-Không tạo đối tượng trực tiếp → chỉ dùng để kế thừa.
-Định nghĩa hành vi chung → nhưng để lớp con quyết định chi tiết thực hiện.
-Tạo chuẩn thống nhất → đảm bảo tất cả lớp con có các phương thức bắt buộc.
