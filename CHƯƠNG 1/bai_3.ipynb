```python
from math import gcd

class PhanSo:
    def __init__(self, tu=0, mau=1):
        if mau == 0:
            raise ValueError("Mẫu số không được bằng 0!")
        self.tu = int(tu)
        self.mau = int(mau)
        self.rut_gon()

    def rut_gon(self):
        g = gcd(self.tu, self.mau)
        if g != 0:
            self.tu //= g
            self.mau //= g
        if self.mau < 0:          
            self.tu = -self.tu
            self.mau = -self.mau

    def __str__(self):
        return f"{self.tu}" if self.mau == 1 else f"{self.tu}/{self.mau}"

    def hien_thi(self):
        print(str(self))

    def __add__(self, other):
        return PhanSo(self.tu * other.mau + other.tu * self.mau, self.mau * other.mau)

    def __sub__(self, other):
        return PhanSo(self.tu * other.mau - other.tu * self.mau, self.mau * other.mau)

    def __mul__(self, other):
        return PhanSo(self.tu * other.tu, self.mau * other.mau)

    def __truediv__(self, other):
        if other.tu == 0:
            raise ZeroDivisionError("Không thể chia cho phân số có tử số = 0")
        return PhanSo(self.tu * other.mau, self.mau * other.tu)
# Ví dụ
if __name__ == "__main__":
    ps1 = PhanSo(2, 4)   
    ps2 = PhanSo(3, 6)  

    print("Phân số 1:", ps1)        
    print("Phân số 2:", ps2)
    print("Tổng:", ps1 + ps2)       
    print("Hiệu:", ps1 - ps2)
    print("Tích:", ps1 * ps2)
    print("Thương:", ps1 / ps2)
```
