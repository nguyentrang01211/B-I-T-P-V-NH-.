| Tiêu chí            | Lập trình cấu trúc (SP)             | Lập trình hướng đối tượng (OOP)        |
| ------------------- | ----------------------------------- | -------------------------------------- |
| Khái niệm           | Chia chương trình thành hàm/thủ tục | Chia chương trình thành lớp/đối tượng  |

| Nguyên lý chính     | Tuần tự, rẽ nhánh, lặp              | Đóng gói, kế thừa, đa hình, trừu tượng |
| Tái sử dụng         | Hạn chế                             | Mạnh nhờ kế thừa, đa hình              |
| Phù hợp             | Chương trình nhỏ, đơn giản          | Hệ thống lớn, phức tạp                 |
