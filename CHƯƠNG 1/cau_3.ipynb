1. Gom dữ liệu và phương thức lại với nhau
-Thuộc tính (data) và phương thức (method) liên quan được gói trong cùng một lớp.
-Đối tượng vừa lưu trữ dữ liệu vừa xử lý dữ liệu đó.
2. Che giấu thông tin (Data hiding)
-Không cho phép truy cập trực tiếp vào dữ liệu nhạy cảm.
-Chỉ có thể thao tác thông qua các phương thức getter/setter hoặc các hàm công khai.
3. Kiểm soát truy cập
-Dùng phạm vi truy cập:
 + Public (công khai)
 + Protected (bảo vệ, chỉ lớp con dùng được)
 + Private (riêng tư, chỉ lớp hiện tại dùng được)
-Giúp ngăn chặn việc thay đổi dữ liệu không mong muốn.
4. An toàn và dễ bảo trì
-Người dùng chỉ cần quan tâm đến giao diện (interface), không cần biết chi tiết cài đặt bên trong.
-Nếu muốn thay đổi cách cài đặt, chỉ sửa bên trong lớp, không ảnh hưởng bên ngoài.
