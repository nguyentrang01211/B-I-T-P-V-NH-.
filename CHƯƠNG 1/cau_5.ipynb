-Đa hình nghĩa là “nhiều hình thái”.
-Trong OOP, đa hình cho phép cùng một tên phương thức/hành động nhưng có cách thực hiện khác nhau tùy đối tượng hoặc ngữ cảnh.
**Đặc trưng
-Cùng tên, hành vi khác: Một phương thức có thể hoạt động khác nhau với từng lớp.
-Linh hoạt: Giúp chương trình dễ mở rộng, dễ thay đổi mà không ảnh hưởng đến phần khác.
-Có 2 dạng phổ biến:
Đa hình lúc biên dịch (Compile-time polymorphism / Overloading)
→ cùng tên hàm nhưng khác tham số.
Đa hình lúc chạy (Runtime polymorphism / Overriding)
→ lớp con định nghĩa lại phương thức của lớp cha theo cách riêng.
