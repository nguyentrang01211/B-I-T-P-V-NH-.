```python
ser1 = pd.Series([1, 2, 3, 4, 5])
ser2 = pd.Series([4, 5, 6, 7, 8])

print("\n--- Bài 4.4 ---")
print("Ser1:", ser1.tolist())
print("Ser2:", ser2.tolist())

# 1. Lấy các phần tử của ser1 không có trong ser2
ser1_only = ser1[~ser1.isin(ser2)]
print("\nCác phần tử của ser1 không có trong ser2:\n", ser1_only)

# 2. Lấy các phần tử chung của ser1 và ser2
common_elements = ser1[ser1.isin(ser2)]
print("\nCác phần tử chung của ser1 và ser2:\n", common_elements)

# 3. Kết hợp hai Series thành một DataFrame (ví dụ minh họa)
df_combined = pd.concat([ser1, ser2], axis=1)
df_combined.columns = ['Series_1', 'Series_2']
print("\nDataFrame kết hợp từ ser1 và ser2:\n", df_combined)
```
