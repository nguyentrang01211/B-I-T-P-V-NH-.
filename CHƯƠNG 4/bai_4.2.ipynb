```python
import pandas as pd
import numpy as np

# Input
mylist = list('abcdefghijklmnopqrstuvwxyz')
myarr = np.arange(26)
mydict = dict(zip(mylist, myarr))
ser = pd.Series(mydict)

# Chuyển đổi Series thành DataFrame
df = ser.to_frame().reset_index()
df.columns = ['index', 'values']  # Đặt tên cột
print(df.head())
```
