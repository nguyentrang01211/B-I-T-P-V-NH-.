```python
import numpy as np
import pandas as pd

# Từ danh sách
mylist = list('abcdefghijklmnopqrstuvwxyz')
series_from_list = pd.Series(mylist)
print("Series từ danh sách:\n", series_from_list)

# Từ mảng numpy
myarr = np.arange(26)
series_from_numpy = pd.Series(myarr)
print("\nSeries từ numpy array:\n", series_from_numpy)

# Từ điển
mydict = dict(zip(mylist, myarr))
series_from_dict = pd.Series(mydict)
print("\nSeries từ điển:\n", series_from_dict)
```
