```python
np.random.seed(100) # Đặt seed để kết quả có thể tái lập
ser_45 = pd.Series(np.random.normal(10, 5, 25))
print("Series cho Bài 4.5:\n", ser_45.head()) # Chỉ in 5 phần tử đầu

min_value_45 = ser_45.min()
percentile_25_45 = ser_45.quantile(0.25)
median_value_45 = ser_45.median()
percentile_75_45 = ser_45.quantile(0.75)
max_value_45 = ser_45.max()
print("Giá trị tối thiểu:", min_value_45)
print("Phần centile thứ 25:", percentile_25_45)
print("Trung vị (Median):", median_value_45)
print("Phần centile thứ 75:", percentile_75_45)
print("Giá trị tối đa:", max_value_45)
```
